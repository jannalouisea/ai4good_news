```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python

# url definition
url = "https://www.cp24.com/news"
```

```python
# Request
r1 = requests.get(url)
r1.status_code

# We'll save in coverpage the cover page content
coverpage = r1.content

# Soup creation
soup1 = BeautifulSoup(coverpage, 'html5lib')

# News identification
coverpage_news = soup1.find_all('h2', class_='teaserTitle')
len(coverpage_news)
```

```python
number_of_articles = len(coverpage_news)
print(number_of_articles)
```

```python
contents = []
links = []
titles = []
body_count = 0

for n in range(number_of_articles):
    link = coverpage_news[n].find('a')['href']
    links.append(link)
    
    title = coverpage_news[n].find('a').get_text()
    titles.append(title)
    
    article = requests.get(link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')
    body = soup_article.find_all('div', class_='articleBody')
    body_count += 1
    print(body_count)
    if(len(body)== 0):
        final_article = " "
    else:
        x=body[0].find_all('p')
    
        #unifying thep aragraphs
        paragraphs = []
        for p in range(len(x)):
            paragraph = x[p].get_text()
            paragraphs.append(paragraph)
            final_article = " ".join(paragraphs)
    
    contents.append(final_article)
```
